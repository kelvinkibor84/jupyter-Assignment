```python
# Task 1: Load and Explore the Dataset
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

# Load Iris dataset
iris = load_iris(as_frame=True)
df = iris.frame

# Add target (species) column
df['species'] = iris.target_names[iris.target]

# Display first few rows
print("First 5 rows of the dataset:")
display(df.head())

# Check data types and missing values
print("\nDataset Info:")
print(df.info())

print("\nMissing values per column:")
print(df.isnull().sum())

# No missing values in Iris dataset, but if there were:
# df = df.dropna() or df.fillna(value)

# --------------------------------------------------

# Task 2: Basic Data Analysis
print("\nBasic Statistics:")
display(df.describe())

# Group by species and calculate mean
species_means = df.groupby('species').mean()
print("\nAverage values per species:")
display(species_means)

# --------------------------------------------------

# Task 3: Data Visualization

# Line Chart - sepal length over index
plt.figure(figsize=(8, 5))
plt.plot(df.index, df['sepal length (cm)'], label='Sepal Length')
plt.title("Sepal Length Trend")
plt.xlabel("Index")
plt.ylabel("Sepal Length (cm)")
plt.legend()
plt.show()

# Bar Chart - average petal length per species
plt.figure(figsize=(8, 5))
sns.barplot(x='species', y='petal length (cm)', data=df, ci=None)
plt.title("Average Petal Length per Species")
plt.xlabel("Species")
plt.ylabel("Petal Length (cm)")
plt.show()

# Histogram - sepal width distribution
plt.figure(figsize=(8, 5))
plt.hist(df['sepal width (cm)'], bins=20, edgecolor="black")
plt.title("Distribution of Sepal Width")
plt.xlabel("Sepal Width (cm)")
plt.ylabel("Frequency")
plt.show()

# Scatter Plot - sepal length vs petal length
plt.figure(figsize=(8, 5))
sns.scatterplot(x='sepal length (cm)', y='petal length (cm)', hue='species', data=df)
plt.title("Sepal Length vs Petal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.legend()
plt.show()

# --------------------------------------------------

# Task 4: Findings and Observations
from IPython.display import Markdown

findings = """
# Findings and Observations

1. The dataset was successfully loaded and cleaned (no missing values remain).
2. Descriptive statistics showed clear differences between numerical features such as sepal length, sepal width, petal length, and petal width.
3. Grouping by species revealed distinct average petal and sepal sizes, confirming the usefulness of these features for classification.
4. Visualizations:
   - **Line chart** showed trends across indices for sepal length.
   - **Bar chart** highlighted average petal length differences between species.
   - **Histogram** displayed the distribution of sepal widths, which was roughly normal.
   - **Scatter plot** showed a strong correlation between petal length and petal width, separating species clearly.
5. Overall, the analysis confirms that the Iris dataset is highly suitable for machine learning tasks such as classification.
"""

display(Markdown(findings))
```
